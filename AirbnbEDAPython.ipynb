Importing the Libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**Data Cleaning**


```python
data.isnull().sum()
```

```python
# Dropping the missing values
data.dropna(inplace = True)
data.shape
```

```python
# Duplicated Records
data.duplicated().sum()
```

```python
#Deleting all duplicate records
data.drop_duplicates(inplace = True)
data.duplicated().sum()
```

**Checking the Data Inconsistencies**

```python
data.dtypes
data['id'] = data['id'].astype(object)
data['host_id'] =data['host_id'].astype(object)
data.dtypes
```


**EDA**


Data Analysis

```python
# Univariate Analysis
data.columns
```

```python
# Identifying outliers in Price
sns.boxplot(data=data, x ='price')
```

```python
# Removing the outliers
data_new = data[data['price']<1500]
sns.boxplot(data=data_new, x='price')
```

```python
# Price Distributions
plt.figure(figsize=(8,5))
sns.histplot(data = data_new , x = 'price',bins =100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
# Price Availability
plt.figure(figsize=(8,5))
sns.histplot(data = data_new , x = 'availability_365')
plt.title("Availability Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
data_new.columns
```

```python
data_new.groupby('neighbourhood_group')['price'].mean()
```

```python
data_new['price_per_bed'] = data_new['price']/data_new['beds']
data_new.head(2)
```

```python
data_new.groupby('neighbourhood_group')['price_per_bed'].mean()
```

Bi Variate Analysis

```python
data.columns
```

```python
# Price dependencies on neighbourhood
sns.barplot(data=data_new, x='neighbourhood_group',y='price',hue='room_type')
```

```python
# Number of Reviews on Neighbourhood
plt.figure(figsize =(10,5))
sns.scatterplot(data=data_new,x='number_of_reviews',y='price',hue = 'neighbourhood_group')
```

```python
# Geographical Distribution of Airbnb Listing
plt.figure(figsize = (12,5))
sns.scatterplot(data = data_new, x='longitude',y='latitude',hue ='room_type')
plt.title('Geographical Distribution of Airbnb Listing')
plt.show()
```

```python
# heat map - Correlation of one variable with others
corr = data_new[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr
```

```python
sns.heatmap(data=corr,annot=True)
```
